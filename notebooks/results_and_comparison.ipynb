```python
import pandas as pd
from sklearn.metrics import classification_report, confusion_matrix
```

```python
baseline_df = pd.read_csv("../data/processed/metrics_baseline.csv")
balanced_df = pd.read_csv("../data/processed/metrics_balanced.csv")
```

```python
baseline_df.head()
balanced_df.head()
```

```python
baseline_df["model"] = "Baseline"
balanced_df["model"] = "Balanced"
```

```python
comparison_df = pd.concat(
    [baseline_df, balanced_df],
    ignore_index=True
)
```

```python
comparison_df = comparison_df[
    ["model", "class", "precision", "recall", "f1_score", "support"]
]
```

```python
comparison_df.to_csv("../data/processed/model_comparison_metrics", index=False)
```
